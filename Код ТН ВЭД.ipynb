```python
import pandas as pd


data = {
    "Код ТН ВЭД": [
        8705908002, 6109100000, 6309000000, 6116930000, 4015120001,
        9001101000, 8414202000, 1101001509, 1102905000, 1102903000
    ],
    "Название товара": [
        "автомобиль", "футболка", "футболка", "перчатки", "перчатки",
        "кабель", "насос", "мука", "мука", "мука"
    ],
    "Описание": [
        "Автомобили грузовые аварийные, бывшие в эксплуатации",
        "Майки, фуфайки с рукавами и прочие нательные фуфайки трикотажные, из хлопчатобумажной пряжи, машинного или ручного вязания",
        "Одежда и прочие изделия, бывшие в употреблении",
        "Прочие перчатки, рукавицы и митенки трикотажные из синтетических нитей машинного или ручного вязания",
        "Перчатки хирургические",
        "Кабели для передачи изображения, волоконно-оптические",
        "Пневматические насосы ручные для велосипедов",
        "Мука из пшеницы мягкой и спельты, прочая",
        "Мука рисовая",
        "Мука овсяная"
    ]
}


tnved_data = pd.DataFrame(data)
```

```python
def classify_product(description, tnved_data):
    for _, row in tnved_data.iterrows():
        if row["Описание"].lower() in description.lower():
            return row["Код ТН ВЭД"]
    return "Код не найден"
```

```python

def verify_code(product_code, tnved_data):
    if product_code in tnved_data["Код ТН ВЭД"].values:
        return "Код корректен"
    return "Ошибка в коде"
```

```python

new_product_description = "Перчатки хирургические"
new_product_code = classify_product(new_product_description, tnved_data)
print(f"Код ТН ВЭД для товара: {new_product_code}")


verification_result = verify_code(new_product_code, tnved_data)
print(f"Результат проверки: {verification_result}")
```
